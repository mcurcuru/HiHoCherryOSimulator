```python
import random
import math

spins = [-1, -2, -3, -4, 2, 2, 10]
total_turns = []

for i in range(100):
    cherrytree = 10
    turns = 0

    while cherrytree > 0:

        spinIndex = random.randrange(0, 7)
        spinResult = spins[spinIndex]

        print ("You spun:", str(spinResult))

        cherrytree += spinResult

        if cherrytree > 10:
            cherrytree = 10
        elif cherrytree < 0:
            cherrytree = 0

        print ("Cherries on your tree:", str(cherrytree))

        turns += 1

    print ("You won the game in",str(turns),"turns!")
    total_turns.append(turns)

avg_turns = sum(total_turns)/len(total_turns)
print ('It took you {} turns on average to win the game.'.format(avg_turns))

ssd = sum((x - avg_turns) ** 2 for x in total_turns)
variance = ssd / (len(total_turns) - 1)
stddev= math.sqrt(variance)
print ('The Standard Deviation for the number of turns to win the game is {}.'.format(stddev))
```

```python
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np
# generating multiple normal distributions
domain = np.linspace(1, 70, 1000) # dividing the distance between -2 and 2 into 1000 points
df = DataFrame(total_turns)
mean = df.mean()
std_value = df.std()

plt.figure(figsize=(16, 9))
for mu, std in zip(mean, std_value):
    # pdf stands for Probability Density Function, which is the plot the probabilities of each range of values
    probabilities = norm.pdf(domain, mu, std)
    plt.plot(domain, probabilities, label=f"$\mu={mu}$\n$\sigma={std}$\n")

plt.legend()
plt.title('Histogram of a Random Sample based on a Normal Distribution')
plt.xlabel("Number of Turns to Win Hi!Ho!Cherry-O!")
plt.ylabel("Probability")
plt.show()
```
